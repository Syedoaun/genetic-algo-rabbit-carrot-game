```python
# Import random module
import random
```

### Initializing Base Variables

```python
# Define the cell types
EMPTY = 0
CARROT = 1
OBSTACLE = 2
TRAP = 3
RABBIT = 4

# Define the action types
UP = 0
DOWN = 1
LEFT = 2
RIGHT = 3

# Define the board size
ROWS = 5
COLS = 5

# Define the initial number of moves and energy for the rabbit
MOVES = 20
ENERGY = 10
```

### User-Defined Functions

```python
# Generate a random board with the given cell types and probabilities
def generate_board(x,y):
    board = []
    for i in range(ROWS):
        row = []
        for j in range(COLS):
            if i == x and j == y:
                cell = RABBIT
                row.append(cell)
            else:
                cell = random.choices([EMPTY, CARROT, OBSTACLE, TRAP], weights=[0.6, 0.2, 0.1, 0.1])[0]
                row.append(cell)
        board.append(row)
    return board
```

```python
def update_board(New_x,New_y,Old_x,Old_y,board):
    board[New_x][New_y] = 4
    board[Old_x][Old_y] = 0
    return board
```

```python
# Display the board in a 2D way with symbols for each cell type
def display_board(board):
    symbols = {EMPTY: ".", CARROT: "C", OBSTACLE: "O", TRAP: "T", RABBIT: "R"}
    for i in range(ROWS):
        for j in range(COLS):
            print(symbols[board[i][j]], end=" ")
        print()
    print()
    
```

```python
# Generate a random initial position and direction for the rabbit
def generate_rabbit():
    x = random.randint(0, ROWS - 1)
    y = random.randint(0, COLS - 1)
    direction = random.randint(0, 3)
    return x, y, direction
```

```python
# Display the rabbit's position, direction, moves, and energy
def display_rabbit(x, y, direction, moves, energy):
    directions = {UP: "↑", DOWN: "↓", LEFT: "←", RIGHT: "→"}
    print(f"Rabbit's position: ({x}, {y})")
    print(f"Rabbit's direction: {directions[direction]}")
    print(f"Rabbit's moves left: {moves}")
    print(f"Rabbit's energy: {energy}")
    print()
```

```python
# Test the code by generating and displaying a board and a rabbit
def Code_Test():
    board = generate_board()
    display_board(board)
    x, y, direction = generate_rabbit()
    display_rabbit(x, y, direction, MOVES, ENERGY)
```

```python
def Size_and_Position_Determination(x,y):
    Size_of_next_population = 0
    Position = "NONE"
    Rabbit_x_Poistion = x
    Rabbit_y_Poistion = y
    
    #Corner Coordinates
    #Top Left Corner
    if Rabbit_x_Poistion and Rabbit_y_Poistion == 0:
        Size_of_next_population = 2
        Position = "Top_LC"
    #Bottom Left Corner
    elif Rabbit_x_Poistion == 4 and Rabbit_y_Poistion == 0:
        Size_of_next_population = 2
        Position = "Bot_LC"
    #Top Right Corner
    elif Rabbit_x_Poistion == 0 and Rabbit_y_Poistion == 4:
        Size_of_next_population = 2
        Position = "Top_RC"
    #Bottom Right Corner
    elif Rabbit_x_Poistion and Rabbit_y_Poistion == 4:
        Size_of_next_population = 2
        Position = "Bot_RC"
        
    #Edge Coordinates
    #Top Edge
    elif Rabbit_x_Poistion == 0 and 0<Rabbit_y_Poistion<4:
        Size_of_next_population = 3
        Position = "Top_E"
    #Bottom Edge
    elif Rabbit_x_Poistion == 4 and 0<Rabbit_y_Poistion<4:
        Size_of_next_population = 3
        Position = "Bot_E"
    #Left Edge
    elif Rabbit_y_Poistion == 0  and 0<Rabbit_x_Poistion<4:
        Size_of_next_population = 3
        Position = "Left_E"
    #Right Edge
    elif Rabbit_y_Poistion == 4 and 0<Rabbit_x_Poistion<4:
        Size_of_next_population = 3
        Position = "Right_C"
        
    #Other Coordinates
    else:
        Size_of_next_population = 4
        Position = "Any"
    
    return Size_of_next_population,Position
    
```
